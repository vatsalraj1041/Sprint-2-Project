Data Cleaning

```python
# City_columns
matches[matches["city"].isna()]
```

```python
matches.loc[matches["city"].isna(),'city'] = 'Dubai'
```

```python
#winner Colums
matches[matches["winner"].isna()]
```

```python
# Winner and Player of the match
matches.loc[matches["winner"].isna(),'winner'] = 'No winner'
matches.loc[matches["player_of_match"].isna(),'player_of_match'] = 'No Player_of_match'
```

```python
#umpire1 and umpire2 columns
matches[matches["umpire1"].isna()]
```

```python
matches.loc[(matches["umpire1"].isna()) | (matches["id"]==5),['umpire1','umpire2']] = ["S Ravi","VK Singh"]
matches.loc[(matches["umpire1"].isna()) | (matches["id"]==11413),['umpire1','umpire2']] = ["Navdeep Singh","Nikhil Patwardhan"]
```

```python
matches.drop("umpire3",axis=1,inplace=True)
```

```python
matches.replace(to_replace = 'Rising Pune Supergiant',value = 'Rising Pune Supergiants',inplace = True)
deliveries.replace(to_replace = 'Rising Pune Supergiant',value = 'Rising Pune Supergiants',inplace = True)

matches.replace(to_replace = 'Delhi Daredevils',value = 'Delhi Capitals',inplace = True)
deliveries.replace(to_replace = 'Delhi Daredevils',value = 'Delhi Capitals',inplace = True)
```
